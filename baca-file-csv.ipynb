# Stopword removal

```python
from nltk.corpus import stopwords
#import nltk
#nltk.download('stopwords')
stopWords = stopwords.words('english')
print(stopWords)
```

# Stemming

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
input_str = "There are several types of stemming algorithms."
input_str = nltk.word_tokenize(input_str)
for word in input_str:
    print(stemmer.stem(word))
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
#import nltk
#nltk.download('omw-1.4')
lemmatizer = WordNetLemmatizer()
input_str = "been had done languages cities mice"
input_str = nltk.word_tokenize(input_str)
for word in input_str:
    print(lemmatizer.lemmatize(word))
```

# Menerapkan Preprocessing pada dataset tiket

```python
data['1'][0:10]
```

```python
lemmatizer = WordNetLemmatizer()
for index, row in data.iterrows():
    filterSentence = []
    sentence = row['1']
    sentence = re.sub(r"[^\w\s]", "", sentence)#cleaning
    words = nltk.word_tokenize(sentence)#tokenization
    words = [w for w in words if not w in stopWords] #stopwords removal
    for word in words:
        filterSentence.append(lemmatizer.lemmatize(word))
    print(filterSentence)
    data.loc[index, '1'] = filterSentence
```
